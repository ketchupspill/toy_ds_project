# **2. Main Question**

The broad question I aim to address: We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts.

My more specific question is: can **experience** and **age** predict how many hours a user will spend playing in the players dataset?

The data will be helpful because I believe there may be some correlation ie greater experience, in a certain age range corresponds more to a certain likeliness of a player playing more.



# **3. Data Analysis and Visualisation**

The following cell tidies the data.

```r
#code to tidy dataset
sessions_tidy <- sessions_untidy |>
                   rename(hashed_email = hashedEmail)
players_tidy  <- players_untidy |>
                   rename(hashed_email = hashedEmail,
                         age = Age)
players     <- players_tidy |>
                   mutate(experience = as_factor(experience),
                          subscribe = as_factor(subscribe),
                          gender = as_factor(gender))

sessions   <- separate(sessions_tidy,
                            col = start_time,
                            into = c("start_date", "start_time"),
                            sep = " ")
sessions    <- separate(sessions,
                            col = end_time,
                            into = c("end_date", "end_time"),
                            sep = " ")
```

```r
#view the first 6 rows of the tidied data.
head(players)
head(sessions)
```

**Summary of Changes**

**Sessions**

- Changed hashedEmail to hashed_email for tidyness.
- Seperated start_time and end_time columns into date and time columns.

**Players**

- Changed experience, gender, and subscribe to factors.
- Renamed hashedEmail and Age to hashed_email and age for tidyness.

```r
#computes means for quantitative variables in the players dataset
mean <- players |>
       select(played_hours, age) |>
       map_df(mean, na.rm = TRUE) |>
       rename(avg_played_hours = played_hours,
              average_age = age)
mean
     
```

# **4. Methods and Plan**

**Method**

We will use k-nearest neighbour model to predict a player's hours based on their experience and age.

**Why is it appropriate?**

It is appropriate because it can be constantly used for each new player as the training model can also become more refined and accurate in its predictions.**

**Assumptions**

Must assume each player actually is at the correct experience level.
Must assume age was truthful.

**Limitations**

This may be too simple of a methodology and could result in unforseen problems if it turns out there isn't a strong correlation between the predictors and the desired result.

**Model steps**

- I will choose a knn model because I think this is better than a linear model, as the predicates are not quantitative but categorical.
- The data will be split 75 training 25 testing.
- Only one split
- I may use cross validation (n=5 for simplicity) if it proves to be more effective than without validation.
